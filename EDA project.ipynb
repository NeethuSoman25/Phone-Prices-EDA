```python
FINDING THE MAXIMUM PRICES EVERY YEAR
```

```python
year=df['announcement_date'].dt.year
ndf=df.groupby(year)['price(USD)'].max().rename_axis('Year')
ndf.to_frame()
print(ndf)
```

```python
fdf=df.loc[(df['announcement_date']>='2020-01-01')]
fdf['brand'].value_counts()
```

```python
DISPLAYING THE HIGHEST PRICES OF EACH BRAND FROM YEAR 2020
```

```python
ndf=df.loc[(df['announcement_date']>='2020-01-01')]
pdf=ndf[['brand','price(USD)']]
pdf=pdf.groupby('brand')
pdf.max('price(USD)')
```

```python
fv = df.loc[(df['announcement_date'] >= '2023-01-01'),['brand','price(USD)']]
fv=fv.groupby('brand').max()
print(fv)
```
